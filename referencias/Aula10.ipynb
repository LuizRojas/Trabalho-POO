## 1. Paradigmas de Programao

Antes de falarmos sobre o paradigma de programao orientado a objeto,  preciso entender *o que so* paradigmas de programao.

Um paradigma de programao **no** se trata de uma linguagem, mas sim, da **forma** de resolver um problema utilizando uma determinada linguagem de programao. Ou seja, um paradigma  um meio, uma metodologia, uma maneira de se classificar uma linguagem de acordo com suas **funcionalidades** ou as estruturas que ela dispe para resoluo de problemas.

Entre os paradigmas existentes, podemos citar o procedural, o funcional, o estruturado, o imperativo e o orientado a objetos. Conhecer os vrios paradigmas de programao, permite abordagens mais objetivas e otimizadas na programao. 

No mundo do desenvolvimento de softwares existem as linguagens conhecidas como multiparadigmas. Como o nome sugere, essas linguagens suportam mais de um paradigma, como  o caso de Python. Um dos paradigmas suportados por ele  o de POO (Programao Orientada a Objetos). Python tambm permite o uso de Programao Funcional, mas isso no ser tratado nessa aula.

 importante notar que, at agora no curso, utilizamos apenas o paradigma imperativo/procedural.




### 1.1. Programao Imperativa

O paradigma de programao imperativa  fortemente baseada no modelo apresentado pelo matemtico *Alan Turing*, denominado "A mquina de Turing", no qual o programador especifica passos que devem ser seguidos em ordem pelo computador.

Quase sempre, em conjunto com a programao imperativa, utilizamos o **paradigma de programao procedural**, na qual o programador descreve as aes, passo-a-passo do que deve ser executado, isto , ele descreve procedimentos.

Esses so os paradigmas que mais utilizamos durante o curso, e os nicos que utilizamos at agora.

*O trecho de cdigo a seguir ilustra um exemplo simples de programao imperativa para calcular a distncia percorrida por um carro:*

````python
# Calculando a distncia percorrida por um carro utilizando programao imperativa:

# DADOS:
velocidade_do_carro = 60 #km/h
tempo_de_viagem = 140 # min

# Processando os dados para obter um resultado:
horas_de_viagem = tempo_de_viagem / 60 #h
distancia_percorrida = velocidade_do_carro * horas_de_viagem #km

# Exibindo a sada:
print(f"{distancia_percorrida:.0f}km")
````

### 1.2. Programao Orientada a Objetos

Originalmente sugerida por *Alan Kay* por volta de 1969 e desenvolvida por outros grandes pesquisadores ao longo do tempo, a **programao orientada a objetos** utiliza conceitos que visam facilitar a legibilidade e manuteno de _softwares_, sem contar com a maior possibilidade de reutilizao do cdigo.

O intuito da sua criao tambm foi o de aproximar o manuseio das estruturas de um programa ao manuseio das coisas do mundo real, da o nome "objeto" como uma algo genrico, que pode representar qualquer coisa tangvel (pessoas, objetos, automveis, empresas, funcionrios, animais, ferramentas, etc). Sendo assim, programao orientada objeto tem como foco principal o item com o qual estamos lidando.

Considerando o exemplo anterior do carro, usando o paradigma de orientao a objetos, teramos um objeto `carro` e tal objeto teria suas **caractersticas especficas**, como por exemplo sua velocidade. Usando a expresso `carro.velocidade = 60` podemos atribuir o valor 60  velocidade do carro. Podemos tambm definir uma classe de objetos denominada `Veiculos` que exibe propriedades comuns a todos os veculos, como por exemplo, velocidade, direo, acelerao, etc. Como mostraremos, isso facilita a criao de carros, caminhes e motocicletas.

### 3.2. Mtodos

Como dito antes, mtodos so as "funes" das classes. Vamos colocar a mo na massa e criar o mtodo `exibe_operacao`:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operao (int ou float)
        elemento2 : segundo elemento da operao (int ou float)
        resultado : resultado da operao (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def exibe_operacao(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        print(string_operacao)
````

````python
soma = Soma(2, 2)
soma.exibe_operacao()
````

Agora ficou muito mais simples exibir a operao. Contudo existem maneiras melhores de implementar essa funcionalidade. A melhor maneira  criar um mtodo chamado `__str__`. Ao exibir um objeto com `print`, _Python_ invoca esse mtodo, que mexe no objeto e nas suas propriedades como se ele fosse uma varivel string, e retorna a string no formato que foi definido dentro da classe.

Assim como `__init__` e `__str__`, voc pode criar alguns mtodos com nomes "especiais" que iro melhorar sua classe, esses mtodos podem ser encontrados [aqui](https://docs.python.org/pt-br/3/library/functions.html).

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operao (int ou float)
        elemento2 : segundo elemento da operao (int ou float)
        resultado : resultado da operao (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        return string_operacao
````

````python
soma = Soma(2, 2)
print(soma)
````

###3.3. Herana

Agora que j vimos como funcionam a criao de uma classe e criao de mtodos. Vejamos como isso pode ser aplicado nos conceitos que vimos de herana.

Utilizaremos agora o exemplo de animais. Vamos criar uma classe chamada `Animal`, que ser a classe "me" da qual as outras herdaro as propriedade e mtodos, sendo que o primeiro ser o mtodo construtor `__init__`.

````python
class Animal:
    """
    Propriedades:
        nome : nome do animal (string)
        cor : nome da cor (string)
        comida : nome do alimento que ele consome (string)
    """
    def __init__(self, nome, cor, comida):
        self.nome = nome
        self.cor = cor
        self.comida = comida
    
    def __str__(self):
        descricao = f"{self.nome}  da cor {self.cor}, e gosta muito de comer {self.comida}"
        return descricao
    
    def comer(self):
      print(f"{self.nome} est comendo {self.comida}.")
    
    def movimento(self):
      print(f"{self.nome} vive correndo por a.")
    
animal = Animal("Bob", "cinza", "alface")
print(animal)
print(animal.nome)
print(animal.comida)
````

Agora, podemos criar as classes de animais especficos que vo herdar essas propriedades. 

````python
class Gato(Animal):
  def __init__(self, nome, cor, comida, qtd_patas): #1
    super().__init__(nome, cor, comida) 
    self.qtd_patas = qtd_patas

gatinho = Gato("Gato", "caramelo", "mingau", 4) #2
print(gatinho) #3
gatinho.movimento() #4
gatinho.comer() #5
print(gatinho.nome) #6
print(gatinho.qtd_patas) #7
````

Observe que no precisamos reescrever todos mtodos novamente para o tipo `Gato`, afinal ele _herdou_ da classe principal todas as propriedades e mtodos. Ns apenas "passamos" a classe `Animal` para ele. 

No entanto, perceba que na classe `__init_` utilizamos o mtodo `super()` para nos referenciar ao mtodo da classe "me", e passamos os trs respectivos parmetros. Alm disso, como queramos que o nmero de patas do gato estivesse informado no problema, acrescentamos ao construtor da classe herdeira a propriedade `qtd_patas`. Ao fazer isso ns estamos _sobrescrevendo_ o construtor que foi herdado.

**OBS:** Note que a classe herdeira, ou seja, a classe derivada, tem acesso a todas as propriedades e mtodos da classe principal. Porm, **o contrrio no  vlido**. A classe "me", no nosso exemplo a classe `Animal`, no tem acesso  propriedade `qtd_patas` da classe `Gato`.

### 5.1. Encapsulamento

Imagine uma classe `Soma` e que essa classe seja composta de 3 propriedades e 2 mtodos, como a seguir:

```
    classe Soma:
        Propriedades:
            Elemento1,
            Elemento2,
            Resultado
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperao()
```

Se quisermos representar a soma $2 + $3 basta executarmos o **construtor** "Soma", como a seguir:

`soma = Soma(2, 3)`

O construtor atribuir  propriedade "Resultado" o valor 5. Entretanto, nada impede que o usurio de nossa classe faa algo como `soma.Resultado = 8` e, ao chamar `soma.exibirOperao()` teramos na sada algo como `2 + 3 = 8`, o que  um absurdo. Como podemos garantir que esse tipo de erro no ocorra?

O encapsulamento de nossas classes est relacionado justamente a isso: evitar modificaes indevidas e manter um determinado nvel de abstrao (uma vez que podemos "esconder" algumas partes de nossa classe, exibindo apenas aquilo que  interessante ao usurio).

Para resolvermos nosso problema com a classe `Soma`, precisamos entender o conceito de **modificadores de acesso**. Isso nada mais  do que uma "trava" para determinados elementos de nossa classe. Propriedades e mtodos podem ser **pblicos** ou **privados**. Como sugerido pela prpria nomenclatura, um elemento privado s pode ser acessado pela prpria classe, sendo seu uso **bloqueado externamente**. Um elemento pblico pode ser acessado e modificado livremente. Em Python todas as propriedades e todos os mtodos so pblicos, _a no ser que voc defina um modificador de acesso_.

escapsulamento.png

<center><b>Figura 4:</b> Exemplo visual de encapsulamento.</center>

Conhecendo tais conceitos, vamos atualizar a representao de nossa classe.

```
    classe Soma:
        Propriedades:
            Elemento1 (Pblica),
            Elemento2 (Pblica),
            Resultado (Pblica)
        Mtodos:
            Soma(Elemento1, Elemento2) (Pblico), # Construtor
            exibirOperao() (Pblico)
```

Logo de cara percebemos que a propriedade `Resultado` no deveria ser livre, tampouco as propriedades `Elemento1` e `Elemento2` (imagine se alterarmos os elementos de uma soma e mantivermos o resultado, na maioria dos casos estaramos "estragando" a soma). Os mtodos, no entanto, devem permanecer pblicos j que  de nosso interesse que o usurio possa acess-los sempre que quiser.

Isso  recorrente em muitas classes, ou seja, os mtodos costumam ser pblicos e as propriedades costumam ser privadas. Vamos atualizar nossa classe, agora sem especificar quando o elemento  pblico:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperao()
```

Mas e se quisermos, por exemplo, apenas realizar um simples `print(soma.Resultado)` ou ento modificar os elementos corretamente?

Para esse fim criaremos mtodos conhecidos como _**getters**_ e _**setters**_. _Getters_ so responsveis por retornar (get) os valores das propriedades e _setters_ so responsveis pela modificao (set) dessas propriedades. Definindo nossos _getters_ e _setters_:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            modificarSoma(Elemento1, Elemento2), # Setter (Define novos valores para Elemento1 e Elemento2)
            receberElementos(), # Retorna os valores de Elemento1 e Elemento2
            receberResultado() # Retorna o valor de Resultado
```

 importante frisar que dentro do mtodo `modificarSoma` no estamos simplesmente modificando `Elemento1` e `Elemento2`. Estamos, na realidade, fazendo algo do tipo:

```
    mtodo modificarSoma(NovoElemento1, NovoElemento2):
        Elemento1 = NovoElemento1
        Elemento2 = NovoElemento2
        Resultado = NovoElemento1 + NovoElemento2
```

Nossa primeira tarefa  definir os `getters` e `setters`. 

Note que j definimos um *setter* (o mtodo `modificar_elementos`), precisamos agora definir os *getters*.

Uma coisa importante que precisamos definir  como chamaremos esses mtodos. Existem duas formas de declarar (e posteriormente chamar) _getters_ e *setters*, observe a tabela abaixo:

    <tr>
        <th></th>
        <th>Jeito no pythnico</th>
        <th>Jeito pythnico</th>
    </tr>
    <tr>
        <td>GET</td>
        <td>get_parametro()</td>
        <td>parametro</td>
    </tr>
    <tr>
        <td>SET</td>
        <td>set_parametro(valor)</td>

Primeiro definiremos do jeito no-_pythnico_ ( mais simples, mas no to elegante):

```python
class Soma:
    [...]
    def get_elemento1(self):
        return self.elemento1
        
    def get_elemento2(self):
        return self.elemento2
        
    def get_resultado(self):
        return self.resultado
```

O jeito _pythnico_ de se fazer isso  utilizando um mecanismo da linguagem chamado ___decorator___, que auxilia o interpretador e pode ser muito til em seu cdigo. 

O _decorator_ basicamente  uma forma prtica e reusvel de **adicionarmos funcionalidades** s nossas funes/mtodos/classes, **sem** precisarmos alterar o cdigo delas. Um __decorator__  implementado como uma funo que _recebe_ uma funo como parmetro, faz algo, ento executa a funo-parmetro e retorna o resultado desta.

No se preocupe se no conseguir entender. Fazer as coisas do jeito no-_pythnico_ no tornar seu cdigo inutilizvel.

Para definir os _getters_ utilizamos o _decorator_ `@property` uma linha antes de definir o mtodo.

O _decorator_ `@property `  utilizado antes de um mtodo quando este ir modificar ou verificar uma propriedade quando  atribudo algum valor a ela ou o seu valor  lido.

```python
class Soma:
    [...]
    @property #l o valor do elemento 1
    def elemento1(self):
        return self.elemento1
        
    @property #le o valor do elemento 2
    def elemento2(self):
        return self.elemento2
        
    @property #l o valor do resultado
    def resultado(self):
        return self.resultado
```

Para os _setters_ utilizamos o _decorator_ `@nome_da_propriedade.setter` (o `@`  resposvel por definir uma _property_):

```python
class Soma:
    [...]
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2
        
    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento2 + valor
```

Desse modo no precisamos mais do mtodo `modificar_elementos`. Vamos ver como ficou nossa classe:

```python

Ainda no podemos usar os recm criados _getters_ e _setters_ (o interpretador apontar erros). Precisamos definir nossas propriedades como privadas. Para isso devemos inserir um _underscore_ duplo (`__`) antes do nome de nossa propriedade. Dessa maneira:

```

Vamos colocar a classe para funcionar:

````python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operao (int ou float)
        __elemento2 : segundo elemento da operao (int ou float)
        __resultado : resultado da operao (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2

    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado

    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
````

A classe abaixo representa um tringulo retngulo. Como bem sabemos um tringulo retngulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento deve seguir a frmula:
$c^2 = a^2 + b^2$

````python
import math
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.cateto_a = cateto_a
        self.cateto_b = cateto_b
        self.hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.cateto_a
    
    @property
    def cateto_b(self):
        return self.cateto_b
    
    @property
    def hipotenusa(self):
        return self.hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.cateto_b = valor
        
TrianguloRetangulo(3, 4)
````

Observe que obtemos uma mensagem de erro do tipo `AttributeError`. Ela nos indica que estamos tentando definir o valor de um atributo sem que o seu `setter` tenha sido definido apropriadamente. Isso ocorreu pois estamos atribuindo valor  propriedade declarada com o decorator `@property` e no  propriedade privada, que nem mesmo chegou a ser declarada. Podemos resolver este problema da seguinte forma:

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variveis para identificar que so privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
````

````python
# Clula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 10
print(tri)
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso  um erro comum ao lidar com objetos e entra na categoria dos erros de semntica, voc no ser notificado pelo interpretador e ele ir aparecer na pior hora possvel. **Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.**

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__atualiza_hipotenusa()
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    #adicionamos aqui a funo para recalcular e atualizar o valor da hipotenusa
    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.cateto_a**2 + self.cateto_b**2)**(1/2)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````

````python
# Clula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 6
tri.cateto_b = 8
print(tri)
````

Isso se deve ao fato de que, em orientao a objetos, erros de semntica tendem a ser mais recorrentes, e mais graves, do que erros de sintaxe.
