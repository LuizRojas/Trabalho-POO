## 1. Paradigmas de Programação

Antes de falarmos sobre o paradigma de programação orientado a objeto, é preciso entender *o que são* paradigmas de programação.

Um paradigma de programação **não** se trata de uma linguagem, mas sim, da **forma** de resolver um problema utilizando uma determinada linguagem de programação. Ou seja, um paradigma é um meio, uma metodologia, uma maneira de se classificar uma linguagem de acordo com suas **funcionalidades** ou as estruturas que ela dispõe para resolução de problemas.

Entre os paradigmas existentes, podemos citar o procedural, o funcional, o estruturado, o imperativo e o orientado a objetos. Conhecer os vários paradigmas de programação, permite abordagens mais objetivas e otimizadas na programação. 

No mundo do desenvolvimento de softwares existem as linguagens conhecidas como multiparadigmas. Como o nome sugere, essas linguagens suportam mais de um paradigma, como é o caso de Python. Um dos paradigmas suportados por ele é o de POO (Programação Orientada a Objetos). Python também permite o uso de Programação Funcional, mas isso não será tratado nessa aula.

É importante notar que, até agora no curso, utilizamos apenas o paradigma imperativo/procedural.




### 1.1. Programação Imperativa

O paradigma de programação imperativa é fortemente baseada no modelo apresentado pelo matemático *Alan Turing*, denominado "A máquina de Turing", no qual o programador especifica passos que devem ser seguidos em ordem pelo computador.

Quase sempre, em conjunto com a programação imperativa, utilizamos o **paradigma de programação procedural**, na qual o programador descreve as ações, passo-a-passo do que deve ser executado, isto é, ele descreve procedimentos.

Esses são os paradigmas que mais utilizamos durante o curso, e os únicos que utilizamos até agora.

*O trecho de código a seguir ilustra um exemplo simples de programação imperativa para calcular a distância percorrida por um carro:*

````python
# Calculando a distância percorrida por um carro utilizando programação imperativa:

# DADOS:
velocidade_do_carro = 60 #km/h
tempo_de_viagem = 140 # min

# Processando os dados para obter um resultado:
horas_de_viagem = tempo_de_viagem / 60 #h
distancia_percorrida = velocidade_do_carro * horas_de_viagem #km

# Exibindo a saída:
print(f"{distancia_percorrida:.0f}km")
````

### 1.2. Programação Orientada a Objetos

Originalmente sugerida por *Alan Kay* por volta de 1969 e desenvolvida por outros grandes pesquisadores ao longo do tempo, a **programação orientada a objetos** utiliza conceitos que visam facilitar a legibilidade e manutenção de _softwares_, sem contar com a maior possibilidade de reutilização do código.

O intuito da sua criação também foi o de aproximar o manuseio das estruturas de um programa ao manuseio das coisas do mundo real, daí o nome "objeto" como uma algo genérico, que pode representar qualquer coisa tangível (pessoas, objetos, automóveis, empresas, funcionários, animais, ferramentas, etc). Sendo assim, programação orientada objeto tem como foco principal o item com o qual estamos lidando.

Considerando o exemplo anterior do carro, usando o paradigma de orientação a objetos, teríamos um objeto `carro` e tal objeto teria suas **características específicas**, como por exemplo sua velocidade. Usando a expressão `carro.velocidade = 60` podemos atribuir o valor 60 à velocidade do carro. Podemos também definir uma classe de objetos denominada `Veiculos` que exibe propriedades comuns a todos os veículos, como por exemplo, velocidade, direção, aceleraçào, etc. Como mostraremos, isso facilita a criação de carros, caminhões e motocicletas.

Alguns recursos da orientação a objetos no curso podem parecer familiares, considerando que já foram mostradas ao longo do curso. Um exemplo é a "função" `lista.append()`. Veremos que a notação de ponto `.` é utilizada para acessar propriedades/atributos de um certo objeto.

###2.1.1. Construtores

O construtor é o método que, como o próprio nome diz, constrói o objeto de uma determinada classe. É semelhante a quando criamos uma variável do tipo int ou string, mas com o diferencial de que agora essa variável terá várias características e funções próprias pertencentes àquele tipo. 

O construtor faz a inicialização do objeto instanciado.

### 2.2. Herança

Ainda em nosso exemplo da classe `Carro`, podemos definir um automóvel como herdeiro da classe `Veículos`. Mas como assim herdeira? Veja bem, se pensarmos nos vários tipos de véiculos (carros, caminhões, motos, vans, etc), todas eles possuem características em comum. Entre elas podemos citar:

* Placa
* Ano
* Combustível
* Cor
* Proprietário

Portanto, podemos definir uma classe mais genérica chamada `Veículo` da seguinte forma:

```
    classe Veículo:
        Propriedades:
            Ano,
            Placa,
            Proprietário 
        Métodos:
            acelerar(),
            ligar_farol(),
            frear(),
            buzinar()
```

Podemos agora recriar a classe `Carro` como **herdeira** da classe `Veículo`:

```
    classe Carro (herdeira de Veículo):
        Métodos:
            abrir_porta(),
            abrir_janela()
```

Perceba que não declaramos nenhuma propriedade. Com a herança, a classe `Carro` **herda** todos os métodos e propriedades da classe `Veículo`. Também é possível adicionar métodos ou propriedades que não estejam definidas nas classe herdada, como é o caso de `carro.abrir_porta()` e `carro.abrir_janela()`.

Caso algum método, tenha o mesmo nome, mas funcione de maneira diferente (como por exemplo uma bicicleta não acende o farol da mesma forma que um carro ou até mesmo uma moto), basta apenas **sobrescrever** os métodos que terão comportamento diferente, isto é, declarar o método novamente dentro da classe, com o mesmo nome da que precisa ser sobrescrita, e declarar as ações que servem àquele objeto.

Perceba o parâmetro `self`. Vamos declarar nossa classe:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
````

### 3.2. Métodos

Como dito antes, métodos são as "funções" das classes. Vamos colocar a mão na massa e criar o método `exibe_operacao`:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def exibe_operacao(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        print(string_operacao)
````

````python
soma = Soma(2, 2)
soma.exibe_operacao()
````

Agora ficou muito mais simples exibir a operação. Contudo existem maneiras melhores de implementar essa funcionalidade. A melhor maneira é criar um método chamado `__str__`. Ao exibir um objeto com `print`, _Python_ invoca esse método, que mexe no objeto e nas suas propriedades como se ele fosse uma variável string, e retorna a string no formato que foi definido dentro da classe.

Assim como `__init__` e `__str__`, você pode criar alguns métodos com nomes "especiais" que irão melhorar sua classe, esses métodos podem ser encontrados [aqui](https://docs.python.org/pt-br/3/library/functions.html).

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        return string_operacao
````

````python
soma = Soma(2, 2)
print(soma)
````

###3.3. Herança

Agora que já vimos como funcionam a criação de uma classe e criação de métodos. Vejamos como isso pode ser aplicado nos conceitos que vimos de herança.

Utilizaremos agora o exemplo de animais. Vamos criar uma classe chamada `Animal`, que será a classe "mãe" da qual as outras herdarão as propriedade e métodos, sendo que o primeiro será o método construtor `__init__`.

````python
class Animal:
    """
    Propriedades:
        nome : nome do animal (string)
        cor : nome da cor (string)
        comida : nome do alimento que ele consome (string)
    """
    def __init__(self, nome, cor, comida):
        self.nome = nome
        self.cor = cor
        self.comida = comida
    
    def __str__(self):
        descricao = f"{self.nome} é da cor {self.cor}, e gosta muito de comer {self.comida}"
        return descricao
    
    def comer(self):
      print(f"{self.nome} está comendo {self.comida}.")
    
    def movimento(self):
      print(f"{self.nome} vive correndo por aí.")
    
animal = Animal("Bob", "cinza", "alface")
print(animal)
print(animal.nome)
print(animal.comida)
````

Agora, podemos criar as classes de animais específicos que vão herdar essas propriedades. 

````python
class Gato(Animal):
  def __init__(self, nome, cor, comida, qtd_patas): #1
    super().__init__(nome, cor, comida) 
    self.qtd_patas = qtd_patas

gatinho = Gato("Gato", "caramelo", "mingau", 4) #2
print(gatinho) #3
gatinho.movimento() #4
gatinho.comer() #5
print(gatinho.nome) #6
print(gatinho.qtd_patas) #7
````

Observe que não precisamos reescrever todos métodos novamente para o tipo `Gato`, afinal ele _herdou_ da classe principal todas as propriedades e métodos. Nós apenas "passamos" a classe `Animal` para ele. 

No entanto, perceba que na classe `__init_` utilizamos o método `super()` para nos referenciar ao método da classe "mãe", e passamos os três respectivos parâmetros. Além disso, como queríamos que o número de patas do gato estivesse informado no problema, acrescentamos ao construtor da classe herdeira a propriedade `qtd_patas`. Ao fazer isso nós estamos _sobrescrevendo_ o construtor que foi herdado.

**OBS:** Note que a classe herdeira, ou seja, a classe derivada, tem acesso a todas as propriedades e métodos da classe principal. Porém, **o contrário não é válido**. A classe "mãe", no nosso exemplo a classe `Animal`, não tem acesso à propriedade `qtd_patas` da classe `Gato`. 
Pense como se fosse uma herança mesmo, a pessoa que recebeu a herança tem acesso a todo o dinheiro da pessoa que morreu. Mas a pessoa que morreu não tem acesso a nada daquele que recebeu.

# 5. Conceitos mais avançados de POO em Python

### 5.1. Encapsulamento

Imagine uma classe `Soma` e que essa classe seja composta de 3 propriedades e 2 métodos, como a seguir:

```
    classe Soma:
        Propriedades:
            Elemento1,
            Elemento2,
            Resultado
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperação()
```

Se quisermos representar a soma $2 + $3 basta executarmos o **construtor** "Soma", como a seguir:

`soma = Soma(2, 3)`

O construtor atribuirá à propriedade "Resultado" o valor 5. Entretanto, nada impede que o usuário de nossa classe faça algo como `soma.Resultado = 8` e, ao chamar `soma.exibirOperação()` teríamos na saída algo como `2 + 3 = 8`, o que é um absurdo. Como podemos garantir que esse tipo de erro não ocorra?

O encapsulamento de nossas classes está relacionado justamente a isso: evitar modificações indevidas e manter um determinado nível de abstração (uma vez que podemos "esconder" algumas partes de nossa classe, exibindo apenas aquilo que é interessante ao usuário).

Para resolvermos nosso problema com a classe `Soma`, precisamos entender o conceito de **modificadores de acesso**. Isso nada mais é do que uma "trava" para determinados elementos de nossa classe. Propriedades e métodos podem ser **públicos** ou **privados**. Como sugerido pela própria nomenclatura, um elemento privado só pode ser acessado pela própria classe, sendo seu uso **bloqueado externamente**. Um elemento público pode ser acessado e modificado livremente. Em Python todas as propriedades e todos os métodos são públicos, _a não ser que você defina um modificador de acesso_.

escapsulamento.png

<center><b>Figura 4:</b> Exemplo visual de encapsulamento.</center>

Conhecendo tais conceitos, vamos atualizar a representação de nossa classe.

```
    classe Soma:
        Propriedades:
            Elemento1 (Pública),
            Elemento2 (Pública),
            Resultado (Pública)
        Métodos:
            Soma(Elemento1, Elemento2) (Público), # Construtor
            exibirOperação() (Público)
```

Logo de cara percebemos que a propriedade `Resultado` não deveria ser livre, tampouco as propriedades `Elemento1` e `Elemento2` (imagine se alterarmos os elementos de uma soma e mantivermos o resultado, na maioria dos casos estaríamos "estragando" a soma). Os métodos, no entanto, devem permanecer públicos já que é de nosso interesse que o usuário possa acessá-los sempre que quiser.

Isso é recorrente em muitas classes, ou seja, os métodos costumam ser públicos e as propriedades costumam ser privadas. Vamos atualizar nossa classe, agora sem especificar quando o elemento é público:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperação()
```

Mas e se quisermos, por exemplo, apenas realizar um simples `print(soma.Resultado)` ou então modificar os elementos corretamente?

Para esse fim criaremos métodos conhecidos como _**getters**_ e _**setters**_. _Getters_ são responsáveis por retornar (get) os valores das propriedades e _setters_ são responsáveis pela modificação (set) dessas propriedades. Definindo nossos _getters_ e _setters_:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            modificarSoma(Elemento1, Elemento2), # Setter (Define novos valores para Elemento1 e Elemento2)
            receberElementos(), # Retorna os valores de Elemento1 e Elemento2
            receberResultado() # Retorna o valor de Resultado
```

É importante frisar que dentro do método `modificarSoma` não estamos simplesmente modificando `Elemento1` e `Elemento2`. Estamos, na realidade, fazendo algo do tipo:

```
    método modificarSoma(NovoElemento1, NovoElemento2):
        Elemento1 = NovoElemento1
        Elemento2 = NovoElemento2
        Resultado = NovoElemento1 + NovoElemento2
```

Nossa primeira tarefa é definir os `getters` e `setters`. 

Note que já definimos um *setter* (o método `modificar_elementos`), precisamos agora definir os *getters*.

Uma coisa importante que precisamos definir é como chamaremos esses métodos. Existem duas formas de declarar (e posteriormente chamar) _getters_ e *setters*, observe a tabela abaixo:

<table>
    <tr>
        <th></th>
        <th>Jeito não pythônico</th>
        <th>Jeito pythônico</th>
    </tr>
    <tr>
        <td>GET</td>
        <td>get_parametro()</td>
        <td>parametro</td>
    </tr>
    <tr>
        <td>SET</td>
        <td>set_parametro(valor)</td>
        <td>parametro=valor</td>
    </tr>
</table>

Primeiro definiremos do jeito não-_pythônico_ (é mais simples, mas não tão elegante):

```python
class Soma:
    [...]
    def get_elemento1(self):
        return self.elemento1
        
    def get_elemento2(self):
        return self.elemento2
        
    def get_resultado(self):
        return self.resultado
```

O jeito _pythônico_ de se fazer isso é utilizando um mecanismo da linguagem chamado ___decorator___, que auxilia o interpretador e pode ser muito útil em seu código. 

O _decorator_ basicamente é uma forma prática e reusável de **adicionarmos funcionalidades** às nossas funções/métodos/classes, **sem** precisarmos alterar o código delas. Um __decorator__ é implementado como uma função que _recebe_ uma função como parâmetro, faz algo, então executa a função-parâmetro e retorna o resultado desta.

Não se preocupe se não conseguir entender. Fazer as coisas do jeito não-_pythônico_ não tornará seu código inutilizável.

Para definir os _getters_ utilizamos o _decorator_ `@property` uma linha antes de definir o método.

O _decorator_ `@property ` é utilizado antes de um método quando este irá modificar ou verificar uma propriedade quando é atribuído algum valor a ela ou o seu valor é lido.

```python
class Soma:
    [...]
    @property #lê o valor do elemento 1
    def elemento1(self):
        return self.elemento1
        
    @property #le o valor do elemento 2
    def elemento2(self):
        return self.elemento2
        
    @property #lê o valor do resultado
    def resultado(self):
        return self.resultado
```

Para os _setters_ utilizamos o _decorator_ `@nome_da_propriedade.setter` (o `@` é resposável por definir uma _property_):

```python
class Soma:
    [...]
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2
        
    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento2 + valor
```

Desse modo não precisamos mais do método `modificar_elementos`. Vamos ver como ficou nossa classe:

```python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.elemento1

    @property
    def elemento2(self):
        return self.elemento2

    @property
    def resultado(self):
        return self.resultado
    
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento1 + valor
```

Ainda não podemos usar os recém criados _getters_ e _setters_ (o interpretador apontará erros). Precisamos definir nossas propriedades como privadas. Para isso devemos inserir um _underscore_ duplo (`__`) antes do nome de nossa propriedade. Dessa maneira:

```python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operação (int ou float)
        __elemento2 : segundo elemento da operação (int ou float)
        __resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado
    
    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
```

Vamos colocar a classe para funcionar:

````python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operação (int ou float)
        __elemento2 : segundo elemento da operação (int ou float)
        __resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2

    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado

    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
````

A classe abaixo representa um triângulo retângulo. Como bem sabemos um triângulo retângulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento deve seguir a fórmula:
$c^2 = a^2 + b^2$

````python
import math
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.cateto_a = cateto_a
        self.cateto_b = cateto_b
        self.hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.cateto_a
    
    @property
    def cateto_b(self):
        return self.cateto_b
    
    @property
    def hipotenusa(self):
        return self.hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.cateto_b = valor
        
TrianguloRetangulo(3, 4)
````

Observe que obtemos uma mensagem de erro do tipo `AttributeError`. Ela nos indica que estamos tentando definir o valor de um atributo sem que o seu `setter` tenha sido definido apropriadamente. Isso ocorreu pois estamos atribuindo valor à propriedade declarada com o decorator `@property` e não à propriedade privada, que nem mesmo chegou a ser declarada. Podemos resolver este problema da seguinte forma:

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variáveis para identificar que são privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 10
print(tri)
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso é um erro comum ao lidar com objetos e entra na categoria dos erros de semântica, você não será notificado pelo interpretador e ele irá aparecer na pior hora possível. **Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.**

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__atualiza_hipotenusa()
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    #adicionamos aqui a função para recalcular e atualizar o valor da hipotenusa
    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.cateto_a**2 + self.cateto_b**2)**(1/2)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 6
tri.cateto_b = 8
print(tri)
````

Agora tudo parece funcionar corretamente. O segredo para a solução de problemas relacionados a classes é sempre ficar atento e lembrar que o depurador é seu melhor amigo. Isso se deve ao fato de que, em orientação a objetos, erros de semântica tendem a ser mais recorrentes, e mais graves, do que erros de sintaxe.
